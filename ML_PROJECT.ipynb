# Handling the imbalanced dataset
# By Oversampling

```python
x= df.drop('Class',axis=1)
y= df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res=SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x_res, y_res, test_size=0.20, random_state=42
)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

```

```python
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred1 = log.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# DecisionTree Classifier

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred2 = dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
recall_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred2)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=50,      # reduce trees
    max_depth=10,         # limit depth
    n_jobs=-1,            # use all CPU cores
    random_state=42
)

rf.fit(x_train, y_train)
```

```python
y_pred3 = rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
precision_score(y_test,y_pred3)
```

```python
recall_score(y_test,y_pred3)
```

```python
f1_score(y_test,y_pred3)
```

# Model Performance Comparison

```python
Final_data = pd.DataFrame({'Models':['LR','DT','RF'],
                           "ACC":[accuracy_score(y_test,y_pred1)*100,
                                  accuracy_score(y_test,y_pred2)*100,
                                  accuracy_score(y_test,y_pred3)*100,
                                  ]})
```

```python
Final_data
```

```python
Final_data.plot(x='Models', y='ACC', kind='bar', legend=False, title='Model Comparison')
```

# Model Evaluation & Analysis

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

models = {
    "Logistic Regression": y_pred1,
    "Decision Tree": y_pred2,
    "Random Forest": y_pred3
}

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, (name, y_pred) in zip(axes, models.items()):
    ConfusionMatrixDisplay.from_predictions(
        y_test,
        y_pred,
        normalize='true',
        cmap='Blues',
        colorbar=False,
        ax=ax
    )
    ax.set_title(name)

fig.suptitle("Normalized Confusion Matrices for All Models", fontsize=14)
plt.tight_layout()
plt.show()
```
